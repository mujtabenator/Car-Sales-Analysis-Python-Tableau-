```python
# Aggregate gender counts
gender_counts = df['Gender'].value_counts()

# Calculate the total annual income for each gender
gender_income = df.groupby('Gender')['Annual Income'].sum()

# function to format the labels to show both percentage and total annual income
def make_autopct(values):
    def my_autopct(pct):
        total = sum(values)
        val = int(round(pct * total / 100.0))
        return '{p:.1f}%\n(${v:d})'.format(p=pct, v=val)
    return my_autopct


# Plot Show

plt.figure(figsize=(10, 10))
plt.pie(gender_counts, labels=gender_counts.index, autopct=make_autopct(gender_income))
plt.title('Gender Distribution with Total Annual Income')
plt.show()
```

```python
# Load Dataset
df = pd.read_csv('/content/Car Sales.xlsx - car_data.csv')

# Calculations
Price_counts = df['Price ($)'].value_counts()



# Calculate the total annual income for each Price
Price_Income = df.groupby('Price ($)')['Annual Income'].sum()
top_5_Models = Price_Income.nlargest(5)

# function to format the labels to show both percentage and total annual income
def make_autopct(values):
    def my_autopct(pct):
        total = sum(values)
        val = int(round(pct * total / 100.0))
        return '{p:.1f}%\n(${v:d})'.format(p=pct, v=val)
    return my_autopct

# Plot Show

plt.figure(figsize=(10, 10))
plt.pie(top_5_Models, labels=top_5_Models.index, autopct=make_autopct(top_5_Models))
plt.title('Price Distribution')
plt.show()
```
